```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("C:\\Users\\GPT BANTWAL\\Downloads\\boston_housing.csv")
df.head(10)
```

```python
x=df.drop('MEDV',axis='columns')
x
```

```python
y=df.MEDV
```

```python
y=df['MEDV']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.33,random_state=0)
    
```

```python
print("xtrain shape:",xtrain.shape)
print("xtest shape:",xtest.shape)
print("ytrain shape:",ytrain.shape)
print("ytest shape:",ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
y_pred = regressor.predict(xtest)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
plt.scatter(ytest, y_pred, c = 'blue')
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.show()
```

```python
train_score=regressor.score(xtrain,ytrain)
test_score=regressor.score(xtest,ytest)
```

```python
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest,y_pred)
print("Mean Square Error : ", mse)
print("Mean Absolute Error : ", mae)
```
